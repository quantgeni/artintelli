# 명사를 중심으로 단어사전 재구축 (2자 이상)

```
##### ------------------ 본 한글주석 by 유형균 (이메일 문의 : quantgeni@gmail.com) ------------------ #####

prgbar = pyprind.ProgBar(len(psent))
posdic = [] # 긍정 사전 정의

for sent in psent:
  words = twitter.pos(sent) # 형태소 분석
  for word in words:
    if (word[1] == 'Noun') & (len(word[0]) >= 2): # 명사만 추출, 2자 이상 // 조건1, 조건2를 &로 연결하되 반드시 괄호를 쳐줄것
      posdic.append(word[0]) # 사전에 추가
  prgbar.update()
 
```

```
##### ------------------ 본 한글주석 by 유형균 (이메일 문의 : quantgeni@gmail.com) ------------------ #####

prgbar = pyprind.ProgBar(len(nsent))
negdic = [] # 부정 사전 정의

for sent in nsent:
  words = twitter.pos(sent) # 형태소 분석
  for word in words:
    if (word[1] == 'Noun') & (len(word[0]) >= 2): # 명사만 추출, 2자 이상 // 조건1, 조건2를 &로 연결하되 반드시 괄호를 쳐줄것
      negdic.append(word[0]) # 사전에 추가
  prgbar.update()
```

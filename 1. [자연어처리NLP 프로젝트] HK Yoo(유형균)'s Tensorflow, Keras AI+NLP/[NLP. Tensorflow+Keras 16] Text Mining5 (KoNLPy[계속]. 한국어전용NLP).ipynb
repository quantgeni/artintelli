```
##### ------------------ 본 한글주석 by 유형균 (이메일 문의 : quantgeni@gmail.com) ------------------ #####

# KoNLP
# NLTK는 영어 정보처리를 위한 패키지
# 반면, KoNLPy는 한국어 정보처리를 위한 패키지임 서울대학교 산업공학과 개발한 형태소 분석기
# 래퍼 패키지인데, 이미 개발된 한글 형태소 분석기를 파이썬에서 바로 사용할 수 있도록 도와줌

# 현재까지 지원하는 형태소 분석기는 모두 5가지
# => Hannanum, Kkma, Komoran, _mecab, Okt

# 사이트 :  konlpy-ko.readthedocs.io

# konlpy설치전 필수 패키지 JPype1 인데 Visual C++ 14.0버전 필요!! => visualstudio
# => 미리 컴파일된 패키지가 배포됨 : www.lfd.uci.edu/~gohlke/pythonlibs
# => JPype1 피키지(JPype1-0.7.0-cp36-cp36m-win_amd64.whl)를 다운로드한 후 터미널에서 설치함
# 설치 : pip install JPype1-0.7.0-cp36-cp36m-win_amd64.whl
# 설치 : pip install konlpy

import konlpy
from konlpy.tag import Okt # 구)Twitter
from konlpy.tag import Kkma

# 트위터 형태소 사전을 사용하기 위해 초기화
twitter = Okt()

txt1 = '아버지가 방에 들어가신다'
txt2 = '나는 보리밥을 먹었다'
txt3 = '롯데마트가 판매하고 있는 흑마늘 양념 치킨이 논란이 되고 있다.'

# 형태소 분석
twitter.pos(txt1) # 중요 품사 기반 추출
twitter.pos(txt2)
twitter.pos(txt3)

twitter.morphs(txt1) # 상세 품사 기반 추출
twitter.morphs(txt2)
twitter.morphs(txt3)

twitter.nouns(txt1) # 명사 기반 추출
twitter.nouns(txt2)
twitter.nouns(txt3)

# konlpy 내장 한국어 말뭉치
# 대한민국 헌법 말뭉치 : kolaw
# 국회법안 말뭉치 : kobill
from konlpy.corpus import kolaw
from konlpy.corpus import kobill

kolaw.fileids()# 말뭉치 이름 출력
kobill.fileids()# 말뭉치 이름 출력

doc1 = kolaw.open('constitution.txt').read()
doc1[:100] # 100자까지 확인
doc2 = kobill.open('1809890.txt').read()
doc2[:100] # 100자까지 확인
```

```
##### ------------------ 본 한글주석 by 유형균 (이메일 문의 : quantgeni@gmail.com) ------------------ #####

# 대한민국 헌법 말뭉치를 이용해서 워드클라우드로 시각화해 봄

import nltk

import numpy as np
from PIL import Image

import matplotlib.pyplot as plt

from wordcloud import WordCloud
from collections import Counter

from nltk import sent_tokenize
from nltk import word_tokenize

from wordcloud import STOPWORDS
from wordcloud import ImageColorGenerator
import matplotlib as mpl


doc1 = kolaw.open('constitution.txt').read()
doc1

words = twitter.nouns(doc1)
wc = Counter(words)
wc = dict(wc.most_common()) # 반드시, 먼저 dict로 변환하여 하기 WordCloud 내 generate_from_frequencies()의 argument로 삽입



fontpath = 'malgun.ttf'

wcimg = WordCloud(font_path=fontpath, 
                  background_color='white', 
                  width=640, height=480).generate_from_frequencies(wc)
                  # .generate_from_text(doc1)

plt.imshow(wcimg, interpolation='bilinear')
plt.axis('off')
plt.show()

# ==> 단어수가 1자인 단어도 추출되어 워드클라우드에 시각화 됨
```

```
##### ------------------ 본 한글주석 by 유형균 (이메일 문의 : quantgeni@gmail.com) ------------------ #####

# 대한민국 헌법 말뭉치를 이용해서 워드클라우드로 시각화
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib as mpl

fontpath = 'malgun.ttf'

# 워드클라우드로 시각화
wcimg = WordCloud(font_path=fontpath, background_color='white', width=640, height=480)\
                  .generate_from_text(doc1)
plt.imshow(wcimg, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```
##### ------------------ 본 한글주석 by 유형균 (이메일 문의 : quantgeni@gmail.com) ------------------ #####

# 추출된 명사들중 2자 이상만 추출해서 워드클라우드로 시각화
tokens = twitter.nouns(doc1)
words = [word for word in tokens if len(word) > 1]
wc = Counter(words)
wc = dict(wc.most_common())

fontpath = 'malgun.ttf'

# 워드클라우드로 시각화
wcimg = WordCloud(font_path=fontpath, background_color='white', width=640, height=480)\
                  .generate_from_frequencies(wc)
                  # .generate_from_text(doc1)
plt.imshow(wcimg, interpolation='bilinear')
plt.axis('off')
plt.show()
```
